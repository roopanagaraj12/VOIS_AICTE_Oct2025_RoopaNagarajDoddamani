```python
missing_percentage = df.isnull().sum() / len(df) * 100


print("Missing Values Percentage Per Column:\n")
print(missing_percentage)
```

```python
print("Shape of the dataset before dropping rows with missing values:")
print(df.shape)

df_dropped_rows = df.dropna()

print("Shape of the dataset after dropping rows with missing values:")
print(df_dropped_rows.shape)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
counts = df["host_identity_verified"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    counts, 
    labels=counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Set3.colors
)

plt.title("Airbnb Listings by host_identity_verified")
plt.show()
```

```python
counts = df["neighbourhood group"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    counts, 
    labels=counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)

plt.title("Airbnb Listings by neighbourhood group")
plt.show()
```

```python
fs = 500  
t = np.linspace(0, 1, fs, endpoint=False)  
signal = np.sin(2*np.pi*50*t) + 0.5*np.sin(2*np.pi*120*t)

plt.figure(figsize=(10,6))
plt.psd(signal, NFFT=512, Fs=fs)   
plt.title("Power Spectrum ")
plt.xlabel("country")
plt.ylabel("Construction year")
plt.grid(True)
plt.show()
```
